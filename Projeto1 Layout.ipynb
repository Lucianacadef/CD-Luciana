## Análise e Resultados

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_csv('DOM2009.csv', sep=',')
dados.UF = dados.UF.astype('category')
dados.V4623 = dados.V4623.astype('category')
dados.V4622 = dados.V4622.astype('category')
dados.V4105 = dados.V4105.astype('category')
dados.UF.cat.categories = ('Rondônia','Acre','Amazonas','Roraima','Pará','Amapá','Tocantins','Maranhão','Piauí','Ceará','Rio Grande do Norte','Paraíba','Pernambuco','Alagoas','Sergipe','Bahia','Minas Gerais','Espírito Santo','Rio de Janeiro','São Paulo','Paraná','Santa Catarina','Rio Grande do Sul','Mato Grosso do Sul','Mato Grosso','Goiás','Distrito Federal')
dados.V4623.cat.categories = ('Seguro','Leve','Moderada', 'Grave', 'Seguro1', 'Leve1', 'Moderada1', 'Grave1')
dados.V4622.cat.categories = ('Sem Redimento','Até 1/4 salário mínimo','Mais de 1/4 até 1/2 de salário mínimo','Mais de 1/2 até 1 salário mínimo','Mais de 1 até 2 salários mínimos','Mais de 2 até 3 salários mínimos','Mais de 3 até 5 salários mínimos','Mais de 5 salários mínimos','Sem declaração')
dados.V4105.cat.categories = ('Urbana','Urbana1','Urbana2', 'Rural', 'Rural1', 'Rural2', 'Rural3', 'Rural4')
dados.V4623[dados.V4623=='Seguro1']='Seguro'
dados.V4623[dados.V4623=='Leve1']='Leve'
dados.V4623[dados.V4623=='Moderada1']='Moderada'
dados.V4623[dados.V4623=='Grave1']='Grave'
dados.V4105[dados.V4105=='Urbana1']='Urbana'
dados.V4105[dados.V4105=='Urbana2']='Urbana'
dados.V4105[dados.V4105=='Urbana1']='Urbana'
dados.V4105[dados.V4105=='Rural1']='Rural'
dados.V4105[dados.V4105=='Rural2']='Rural'
dados.V4105[dados.V4105=='Rural3']='Rural'
dados.V4105[dados.V4105=='Rural4']='Rural'
DFALL = pd.DataFrame({'Unidade Federal':dados.UF,"Insegurança Alimentar":dados.V4623, "Situação censitária":dados.V4105, "Rendimento Mensal Per Capta":dados.V4622})
DF2 = DFALL.loc[:,['Unidade Federal',"Situação censitária","Rendimento Mensal Per Capta","Insegurança Alimentar"]]
DF2
```

```python
df_sem_nulo = DF2[pd.notnull(DF2["Situação censitária"])]
df_sem_nulo = DF2[pd.notnull(DF2["Rendimento Mensal Per Capta"])]

```

Para uma análise mais eficiente da tabela, todos os dados foram filtrados (retirada dos NaN, Null e brancos). Portanto, um quinto dos dados foram desconsiderados. Vale ressaltar que a análise dos dados foi feita de forma geral, e não proporcionalmente ao número de habitantes do Estado.

```python
df_sem_nulo
```

 O gráfico gerado a seguir é uma correlação da insegurança alimentar por estado. O intuito é descobrir qual é o estado com o maior número de insegurança alimentar grave.

```python

seguranca_alimentar_estados = pd.crosstab(dados.UF, dados.V4623)
seguranca_alimentar_estados.plot(kind='bar', color=('green', 'yellow', 'blue', 'red', 'green', 'yellow', 'blue', 'red'), figsize=(15,5))
plt.ylabel('Insegurança alimentar')
```

A partir deste foi possível notar que o estado com o maior grau de insegurança alimentar grave é a Bahia. A seguir relacionaremos o estado em questão com o rendimento mensal per capita e situação censitária (moradores em área rural ou urbana), a fim de descobrir se existe alguma relação entre essas variáveis.

```python
dados[dados.UF=="Bahia"].V4105.value_counts().plot(kind="pie",autopct='%0.2f')
```

```python
dados[dados.UF=="Bahia"].V4622.value_counts().plot(kind="pie")
```

Através da análise dos gráficos gerados é possível concluir que a maior parte da população da Bahia vive em áreas urbanas, e, praticamente, metade da população possui faixas de renda salarial de 1/4 à 1 salário mínimo.

A partir desses dados iniciei uma pesquisa mais a fundo sobre o assunto, e pude constatar que de 2009 à 2013 a Bahia perdeu a primeira posição de estado com o maior nível de insegurança alimentar grave para o Maranhão, Piauí e Amazonas. O dado mais relevante, nesse caso, para julgar tal condição é a situação censitária em que o estado se apresenta. Contudo, vale destacar que a Bahia é o sexto estado com a menor média de renda do país. Portanto, o rendimento mensal per capita é de grande importância para uma análise a parte da situação em que a Bahia se encontra.

A partir de agora faremos um análise baseada somente na segurança alimentar do estado somada com a situação censitária, de acordo com os dados que foram adquiridos por meio de uma pesquisa independente, uma vez que a PNAD de 2013 fornecida não possui dados suficientes que nos levem a qualquer conclusão.
